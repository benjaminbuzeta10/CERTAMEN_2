```python
df=dt.copy()

import numpy as np
import pandas as pd
import plotly.express as px

# Asegurar columnas requeridas
for col in ["Age", "Purchase Amount (USD)"]:
    if col not in dt.columns:
        raise KeyError(f"Falta la columna '{col}' en el DataFrame.")

# Crear grupos de edad por dÃ©cadas
age_min = max(10, int(dt["Age"].min()//10*10))
age_max = min(80, int(np.ceil(dt["Age"].max()/10)*10))
bins = list(range(age_min, age_max+10, 10)) if age_max > age_min else [10,20,30,40,50,60,70,80]

# Crear etiquetas personalizadas para evitar objetos Interval
labels = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]
dt["AgeGroup"] = pd.cut(dt["Age"], bins=bins, labels=labels, right=False)

# Elegir columna de mÃ©todo de pago si existe
pay_col = "Payment Method" if "Payment Method" in dt.columns else None
if pay_col is None:
    raise KeyError("No encontrÃ© la columna 'Payment Method' para el heatmap.")

# Tabla dinÃ¡mica de promedios con observed=False para silenciar el warning
pivot_mean = (
    dt.pivot_table(
        index="AgeGroup",
        columns=pay_col,
        values="Purchase Amount (USD)",
        aggfunc="mean",
        observed=False  # Agregar este parÃ¡metro
    ).sort_index()
)

fig = px.imshow(
    pivot_mean,
    text_auto=".1f",
    color_continuous_scale="YlGnBu",
    aspect="auto",
    title="Promedio de compra por Grupo de Edad Ã— MÃ©todo de Pago"
)
fig.update_layout(xaxis_title="MÃ©todo de Pago", yaxis_title="Grupo de Edad", coloraxis_colorbar_title="USD")
fig.show()
```

```python
df=dt.copy()

import numpy as np, pandas as pd
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import plotly.express as px

AMOUNT = "Purchase Amount (USD)"
assert AMOUNT in df.columns, f"Falta columna '{AMOUNT}'"

has_pay = "Payment Method" in df.columns
has_season = "Season" in df.columns
has_gender = "Gender" in df.columns

# Grupos de edad
age_min = max(10, int(df["Age"].min() // 10 * 10))
age_max = min(80, int(np.ceil(df["Age"].max() / 10) * 10))
bins = list(range(age_min, age_max + 10, 10))
df_dash = df.copy()
df_dash["AgeGroup"] = pd.cut(df_dash["Age"], bins=bins, right=False)
df_dash["AgeGroupStr"] = df_dash["AgeGroup"].astype(str)

# Dropdown options
pay_opts = sorted(df_dash["Payment Method"].dropna().unique()) if has_pay else []
season_opts = sorted(df_dash["Season"].dropna().unique()) if has_season else []
gender_opts = sorted(df_dash["Gender"].dropna().unique()) if has_gender else []

# Layout
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H3("Dashboard â€” Preferencias de Compra"),
    html.Div([
        dcc.Dropdown(id="pay_dd",
                     options=[{"label":p,"value":p} for p in pay_opts],
                     value=pay_opts[:3] if pay_opts else None,
                     multi=True, placeholder="MÃ©todo(s) de pago"),
        dcc.Dropdown(id="season_dd",
                     options=[{"label":s,"value":s} for s in season_opts],
                     value=season_opts if season_opts else None,
                     multi=True, placeholder="Temporada") if has_season else html.Div(),
        dcc.Dropdown(id="gender_dd",
                     options=[{"label":g,"value":g} for g in gender_opts],
                     value=gender_opts if gender_opts else None,
                     multi=True, placeholder="GÃ©nero") if has_gender else html.Div(),
    ], style={"display":"flex","gap":"10px"}),

    dcc.Tabs([
        dcc.Tab(label="Heatmap Edad Ã— MÃ©todo", children=[dcc.Graph(id="heatmap")]),
        dcc.Tab(label="Tendencias por Edad", children=[dcc.Graph(id="line_agegroup")]),
        dcc.Tab(label="DistribuciÃ³n por MÃ©todo", children=[dcc.Graph(id="box_payment")]),
    ]),
])

def filtrar(dfin, pay_sel, season_sel, gender_sel):
    d = dfin.copy()
    if has_pay and pay_sel: d = d[d["Payment Method"].isin(pay_sel)]
    if has_season and season_sel: d = d[d["Season"].isin(season_sel)]
    if has_gender and gender_sel: d = d[d["Gender"].isin(gender_sel)]
    return d

@app.callback(
    Output("heatmap","figure"),
    Output("line_agegroup","figure"),
    Output("box_payment","figure"),
    Input("pay_dd","value"),
    Input("season_dd","value") if has_season else Input("pay_dd","value"),
    Input("gender_dd","value") if has_gender else Input("pay_dd","value"),
)
def actualizar(pay_sel, season_sel=None, gender_sel=None):
    d = filtrar(df_dash, pay_sel, season_sel, gender_sel)

    # Heatmap
    if has_pay and not d.empty:
        piv = d.pivot_table(index="AgeGroupStr", columns="Payment Method",
                            values=AMOUNT, aggfunc="mean", observed=False)
        fig_h = px.imshow(piv, text_auto=".1f", color_continuous_scale="YlGnBu",
                          title="Promedio de compra por Grupo de Edad Ã— MÃ©todo")
    else:
        fig_h = px.imshow([[0]], text_auto=True, title="Sin datos")

    # LÃ­nea
    if has_pay and not d.empty:
        g = d.groupby(["AgeGroupStr","Payment Method"], dropna=False)[AMOUNT].mean().reset_index()
        fig_l = px.line(g, x="AgeGroupStr", y=AMOUNT, color="Payment Method", markers=True,
                        title="Tendencia por Grupo de Edad y MÃ©todo")
    else:
        fig_l = px.line(title="Sin datos")

    # Boxplot
    if has_pay and not d.empty:
        color_col = "Gender" if has_gender else None
        fig_b = px.box(d, x="Payment Method", y=AMOUNT, color=color_col,
                       points="suspectedoutliers",
                       title="DistribuciÃ³n por MÃ©todo de Pago" + (" y GÃ©nero" if color_col else ""))
    else:
        fig_b = px.box(title="Sin datos")

    return fig_h, fig_l, fig_b

app.run_server(mode="inline", port=8052)
```
